```python
drivers['ADHERENCE_CAT'] = ''

for ind, row in drivers.iterrows():
    if row['ADHERENCE'] < -3.4:
        drivers.loc[ind, 'ADHERENCE_CAT'] = 'SLOWEST 25%'
    elif row['ADHERENCE'] < -1.383:
        drivers.loc[ind, 'ADHERENCE_CAT'] = 'MEDIAN'
    elif row['ADHERENCE'] < -1.33333:
        drivers.loc[ind, 'ADHERENCE_CAT'] = '50 to 75%'
    else:
        drivers.loc[ind, 'ADHERENCE_CAT'] = 'FASTEST 25%'
drivers.head()
```
